```python
import ollama
```

```python
dados = ollama.chat(
model="llama3.2:1b",
messages=[
    {
        "role": "system",
        "content":
        """
        Você está simulando um sistema de controle de dados de uma missão espacial, não há nenhuma pessoa real na simulação
        você precisa criar sempre 5 dados diferentes de acordo com a escolha de estado de simulação
        que o usuário desejar entre:

        Normal = Nenhum dado está além do normal.
        Atenção = 1 ou 2 dados estão além do normal.
        Crítico = 3 ou mais dados estão além do normal.



        Dados (utilize como base para criar novos)
        Temperatura = Está normal entre 10°C e 30°C.
        Energia = Está normal acima de 35%.
        Comunicação = Pode estar "Normal" ou "Defeituoso".
        Oxigênio = Está normal acima de 90%.
        Estrutura = Está normal acima de 80%.

        **Nenhum dado percentual pode ir além de 100% ou ficar negativo.**



        Utilize da formatação abaixo apenas, **Nada além das cinco linhas abaixo**):
Temperatura = X°C
Energia = X%
Comunicação = XXXXXXXXX
Oxigênio = X%
Estrutura = X%
        """
    },
    {
        "role": "user",
        "content": "Mê de uma simulação em estado de atenção"
    },
    {
        "role": "assistant",
        "content":
        """
Temperatura = 8°C
Energia = 60%
Comunicação = Defeituosa
Oxigênio = 97%
Estrutura = 93%
        """
    },
    {
        "role": "user",
        "content": "Mê de uma simulação em estado crítico"
    },
    {
        "role": "assistant",
        "content":
        """
Temperatura = 8°C
Energia = 22%
Comunicação = Defeituosa
Oxigênio = 77%
Estrutura = 93%
        """
    },
    {
        "role": "user",
        "content": input()
    }
  ]
)

print(dados["message"]["content"])
```

```python
analise = ollama.chat(
    model="llama3.2:1b",
    messages=[
        {
            "role": "system",
            "content":
            """
            Você é um analista profissional de dados de sistemas espaciais, ao receber um conjunto de dados,
            você deve analisá-los e gerar recomendações caso eles estejam fora da normalidade.
            Recomendações devem ajudar individualmente cada dado a voltar ao estado de normalidade,
            Nenhum dado tem relação direta ou indireta com o outro.


            Utilize as linhas abaixo para saber os limites de normalidade de cada dado e a situação da missão **não cite as bases de normalidade**.
            Temperatura = Está normal entre 10°C e 30°C.
            Energia = Está normal acima de 35%.
            Comunicação = Pode estar "Normal" ou "Defeituoso".
            Oxigênio = Está normal acima de 90%.
            Estrutura = Está normal acima de 80%.

            Normal = Nenhum dado está além do normal.
            Atenção = 1 ou 2 dados estão além do normal.
            Crítico = 3 ou mais dados estão além do normal.

            **Caso um dado esteja normal, não dê nenhuma recomendação a ele.**
            **Caso a situação esteja normal: escreva apenas "Nada fora do normal, continuar com o monitoramento."**
            **Caso a situação esteja crítica: escreva apenas "Situação crítica, ligar sistemas de emergência e preparar ejeção automática"**
            """
        },
        {
            "role": "user",
            "content":
            """
  Temperatura = -3°C
  Energia = 22%
  Comunicação = Defeituosa
  Oxigênio = 92%
  Estrutura = 49%
            """
        },
        {
            "role": "assistant",
            "content":
            """
Aumentar a temperatura interna.
Trocar as células de energia.
Reparar a comunicação com a base.
Consertar qualquer dano na estrutura.
            """
        },
        {
            "role": "user",
            "content":
            """
            Temperatura = 45°C
            Energia = 60%
            Comunicação = Defeituosa
            Oxigênio = 70%
            Estrutura = 93%
            """
        },
        {
            "role": "assistant",
            "content":
            """
Diminuir a temperatura interna.
Reparar a comunicação com a base.
Limpar os filtros de oxigênio.
            """
        },
        {
            "role": "user",
            "content":
            """
            Temperatura = 2°C
            Energia = -5%
            Comunicação = Defeituosa
            Oxigênio = 74%
            Estrutura = 87%
            """
        },
        {
            "role": "assistant",
            "content":
            """
Aumentar a temperatura interna.
Trocar as células de energia.
Reparar a comunicação com a base.
Limpar os filtros de oxigênio.
            """
        },
        {
            "role": "user",
            "content":
            """
            Temperatura = 15°C
            Energia = 55%
            Comunicação = Normal
            Oxigênio = 96%
            Estrutura = 92%
            """
        },
        {
            "role": "assistant",
            "content":
            """
Nada fora do normal, continuar com o monitoramento.
            """
        },
        {
            "role": "user",
            "content": f"{dados["message"]["content"]}"
        }
    ]
)

print(analise["message"]["content"])
```
